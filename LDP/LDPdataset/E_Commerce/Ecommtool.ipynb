```python
import LDP.basicFunction.basicfunc as bf
k_path='../E_Commerce/data/origin_Ecomm_k.txt'
v_path = '../E_Commerce/data/origin_Ecomm_v.txt'
lable_path = '../E_Commerce/data/Ecomm_lable.txt'

data_k = bf.readtxt(k_path)
data_v = bf.readtxt(v_path)
data_l = bf.readtxt(lable_path)
l=len(data_l[0])
n=len(data_k)
```

```python
#构建映射字典
d1=data_l[0]
d2=[i+1 for i in range(len(data_l[0]))]

yinshe= dict(zip(d1, d2))
yinshe
```

```python
# 根据字典的映射替换list中的值
def replace(list,dict):
    res=[dict[i] if i in dict else i for i in list]
    return res
```

```python
# 将k中的项id替换成标准编号（从1开始）
k=[]
for i in range(n):
    k.append(replace(data_k[i],yinshe))
k
```

```python
# 归一化为[-1,1]函数
# 保留一位小数，要不然github传不上去
def normal(data,maxium,minium):
    y=2*(data-maxium)/(maxium-minium)+1
    return float('{:.3f}'.format(y))
```

```python
# 将v归一化到[-1,1]
# v中最大值为5，最小值为1
v=[]
maxium=max(data_v)[0]
minium=min(data_v)[0]
for i in range(n):
    tmp=[normal(x,maxium,minium) for x in data_v[i]]
    v.append(tmp)
v
    
```

```python
# 保存标准化之后的v
bf.savetxt(v,'../E_Commerce/data/Ecomm_v.txt')
```
