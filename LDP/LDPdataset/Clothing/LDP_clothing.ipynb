```python
import matplotlib

matplotlib.use('TkAgg')
import LDP.basicDP.OLHbasic as lhb
import LDP.basicFunction.basicfunc as bf
import LDP.basicDP.LDPKVbasic as lkvb
import numpy as np
import random
import time

"""数据处理"""
# 关闭科学计数法显示
np.set_printoptions(suppress=True)
```

#### 读取文件
#### 构建kv对元组列表，每一行代表一个用户的kv对list

```python
k_path='../Clothing/data/Clothing_k.txt'
v_path = '../Clothing/data/Clothing_v.txt'
lable_path = '../Clothing/data/Clothing_lable.txt'

data_k = bf.readtxt(k_path)
data_v = bf.readtxt(v_path)
data_l = bf.readtxt(lable_path)

n = len(data_k)  # 用户数量
l = len(data_l[0])  # key域的长度
epsilon = 5

# 构建kv对元组列表，每一行代表一个用户的kv对list
kv = []

for i in range(n):
    data_k[i]=list(map(int,data_k[i]))
    tmp = zip(data_k[i], data_v[i])
    kv.append(list(tmp))

#label = list(map(int, data_l[0]))  # 查找数据用的label.（不需要再用这个了）！！
    
# 构建label标签，我已经把标签规范化为[1,l]的数字了，也就是说不需要用之前的原始标签了
label=[i+1 for i in range(l)]
```

## 1.采样
#### 注意！！！！要转换为和samplek相同的格式，因为hash是专为str计算的，int和float转换得到的结果不同

```python
samplekv = []
for i in range(n):
    random.seed(1)
    tmp = random.sample(kv[i], 1)
    samplekv.append(tmp[0])
samplekv_unzip = list(zip(*samplekv))
samplek = samplekv_unzip[0]
#samplev = samplekv_unzip[1]

#这里转为dict，速度会快指数倍  !!!!不能转，因为转了之后相同的k值只会出现一次，少了很多值！！！！
# samplekv_dict=dict(samplekv)
# samplek=list(samplekv_dict.keys())
# 
# #samplev=list(samplekv_dict.values())
samplek


```

## 2.用OLH查找候选集

```python
# starttime = time.clock()
# print("开始执行OLH...")
# 
# # 调用OLH
# est = lhb.OLH(epsilon, samplek, n, label)
# 
# # 程序运行结束时间
# endtime = time.clock()
# print('运行时间:', endtime - starttime)
# 
# est

starttime = time.clock()
print("开始执行OLH...")

# 调用OLH
est=lhb.OLH_1(epsilon,samplek,n,l)

# 程序运行结束时间
endtime = time.clock()
print('运行时间:', endtime - starttime)

est
```

```python
# 保存采样结果
bf.savetxt(est,'../Clothing/result/sampleresult.txt')
```

#### 生成top_k的候选集(候选集长度为2k)

```python
k=20
est.sort(key=lambda x:x[1],reverse=True)
topk_candi=est[:2*k]

# candi_unzip = list(zip(*topk_candi))
# candi_k = list(candi_unzip[0])

#在不会出现相同key的情况下，可以转为dict，提升速度
candi_dict=dict(topk_candi)
candi_k=list(candi_dict.keys())
candi_k
```

```python
# 保存候选集，长度为2k
bf.savetxt(candi_k,'../Clothing/result/candidate.txt')
```

#### 与每个用户的项集kv取交集得到用户本地项集minikv

```python
minik=[]
miniv=[]
minikv=[]
# 每个用户的交集项个数
numsize=[]
for i in range(n):
    # data_k[i]是每个用户的项集的k值
    # 将data_k[i]与候选集candi_k取交集得到tmpk
    tmpk=list(set(candi_k).intersection(set(data_k[i])))
    # 根据tmpk得到对应项的v值tmpv
    di=dict(kv[i])
    tmpv=list(map(di.get,tmpk))
    minik.append(tmpk)
    miniv.append(tmpv)
    minikv.append(list(zip(tmpk,tmpv)))
    numsize.append(len(tmpk))
    
numsize
```

```python
# 保存用户取交集之后的本地项集minik和miniv
bf.savetxt(minik,'../Clothing/result/minik.txt')
bf.savetxt(miniv,'../Clothing/result/miniv.txt')
```

## 3.用OLH上传交集的项数

```python
starttime = time.clock()
print("开始执行OLH...")
numsize_lable=[i for i in range(2*k+1)]
# 调用OLH
numsize_est = lhb.OLH(epsilon, numsize, n, numsize_lable)

# 程序运行结束时间
endtime = time.clock()
print('运行时间:', endtime - starttime)

numsize_est
```
