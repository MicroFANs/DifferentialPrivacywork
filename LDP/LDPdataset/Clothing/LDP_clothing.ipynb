```python
import matplotlib

matplotlib.use('TkAgg')
import LDP.basicDP.RPbasic as rpb
import LDP.basicDP.SHbasic as shb
import LDP.basicDP.OLHbasic as lhb
import LDP.basicFunction.basicfunc as bf
import numpy as np
import random
import time

"""数据处理"""
# 关闭科学计数法显示
np.set_printoptions(suppress=True)

```

#### 读取文件
#### 构建kv对元组列表，每一行代表一个用户的kv对list

```python
k_path='../Clothing/data/Clothing_k.txt'
v_path = '../Clothing/data/Clothing_v.txt'
lable_path = '../Clothing/data/Clothing_lable.txt'

data_k = bf.readtxt(k_path)
data_v = bf.readtxt(v_path)
data_l = bf.readtxt(lable_path)

n = len(data_k)  # 用户数量
l = len(data_l[0])  # key域的长度
epsilon = 5

# 构建kv对元组列表，每一行代表一个用户的kv对list
kv = []
for i in range(n):
    tmp = zip(data_k[i], data_v[i])
    kv.append(list(tmp))
```

## 1.采样
#### 注意！！！！要转换为和upk相同的格式，因为hash是专为str计算的，int和float转换得到的结果不同

```python
upkv = []
for i in range(n):
    random.seed(1)
    tmp = random.sample(kv[i], 1)
    upkv.append(tmp[0])
# print(upkv)
upkv_unzip = list(zip(*upkv))
upk = upkv_unzip[0]
upv = upkv_unzip[1]
upk = list(map(int, upk))

label = list(map(int, data_l[0]))  # 查找数据用的label
```

#### upk的真实计数结果

```python
count_true = []  # 真实计数结果
for i in range(l):
    count_true.append((label[i], upk.count(label[i])))

count_true
```

```python
upkpath='../Clothing/result/upkcount.txt'
bf.savetxt(count_true,upkpath)
```

## 2.用OLH查找候选集

```python
starttime = time.clock()
print("开始执行OLH...")

# 调用OLH
est = lhb.OLH(epsilon, upk, n, label)

# 程序运行结束时间
endtime = time.clock()
print('运行时间:', endtime - starttime)

est
```

```python
# 保存候选集
bf.savetxt(est,'../Clothing/result/OLHresult.txt')
```

#### 生成top_k的候选集

```python
k=40
est.sort(key=lambda x:x[1],reverse=True)
est
topk_candi=est[:k]
bf.savetxt(topk_candi,'../Clothing/result/candidate.txt')
```
