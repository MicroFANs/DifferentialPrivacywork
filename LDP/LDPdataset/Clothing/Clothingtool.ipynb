```python
import LDP.basicFunction.basicfunc as bf
k_path='../Clothing/data/origin_Clothing_k.txt'
v_path = '../Clothing/data/origin_Clothing_v.txt'
lable_path = '../Clothing/data/Clothing_lable.txt'

data_k = bf.readtxt(k_path)
data_v = bf.readtxt(v_path)
data_l = bf.readtxt(lable_path)
l=len(data_l[0])
n=len(data_k)
```

```python
#构建映射字典
d1=data_l[0]
d2=[i+1 for i in range(len(data_l[0]))]

yinshe= dict(zip(d1, d2))
yinshe
```

```python
# 根据字典的映射替换list中的值
def replace(list,dict):
    res=[dict[i] if i in dict else i for i in list]
    return res
```

```python
# 将k中的项id替换成标准编号（从1开始）
k=[]
for i in range(n):
    k.append(replace(data_k[i],yinshe))
k
```

```python
# 保存标准化之后的k
#bf.savetxt(k,'../Clothing/data/Clothing_k.txt')

```

```python
# 归一化为[-1,1]函数
def normal(data,maxium,minium):
    y=2*(data-maxium)/(maxium-minium)+1
    return float('{:.3f}'.format(y))
```

```python
# 将v归一化到[-1,1]
# v中最大值为10，最小值为0
v=[]
for i in range(n):
    tmp=[normal(x,10,0) for x in data_v[i]]
    v.append(tmp)
v
    
```
